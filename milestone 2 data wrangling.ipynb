```python
#convert date from object to date format
# 
df['date'] = pd.to_datetime(df['date'] )
df = df.set_index(['date'])
df.sort_index(inplace = True, ascending = True)
df.loc['2020-01-13':'2021-03-07']

```

```python
# dropping unnecessary columns
df = df[['location','total_vaccinations', 'total_distributed','people_vaccinated','people_fully_vaccinated']] 
df
```

```python
df1 = df.groupby(['location'])['total_vaccinations', 'total_distributed','people_vaccinated'].sum().reset_index()
```

```python
df1.isnull().sum()
df1 = df1.fillna(0)
```
